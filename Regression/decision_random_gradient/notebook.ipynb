```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

     

np.random.seed(42)
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
```

```python

plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score


dt = DecisionTreeRegressor(max_depth = 10,random_state=42)
dt.fit(X_train , y_train)

y_pred_dt = dt.predict(X_test)

print(f'r2 score: {r2_score(y_test,y_pred_dt)}')
```

```python


# Create mesh grid
x_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 100)

xx1, xx2 = np.meshgrid(x_range, y_range)

# Predict using FITTED model
y_hat = dt.predict(np.c_[xx1.ravel(), xx2.ravel()])
y_hat = y_hat.reshape(xx1.shape)

# Plot
plt.figure(figsize=(6, 6))
plt.contourf(xx1, xx2, y_hat, alpha=0.3, cmap="viridis")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis")
plt.title("Decision Tree Regression Surface (Tuned)")
plt.show()
```

# Random Forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {
    "n_estimators": [200, 400,500],
    "max_depth": [None, 5,10, 20],
    "min_samples_leaf": [1, 2,3,4,5],
    "max_features": ["sqrt", "log2"]
}

grid = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

grid.fit(X_train, y_train)

print(grid.best_params_)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 500,max_depth= 10, max_features= 'sqrt', min_samples_leaf=5)
rf.fit(X_train , y_train)

y_pred_rf = rf.predict(X_test)

print(f'r2 score: {r2_score(y_test,y_pred_rf)}')
```

```python
plt.figure(figsize=(6, 6))
x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, x_range)
y_hat = rf.predict(np.c_[xx1.ravel(), xx2.ravel()])
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, xx2, y_hat, alpha=0.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha=.7)
plt.title("Random forest")
plt.show()
```

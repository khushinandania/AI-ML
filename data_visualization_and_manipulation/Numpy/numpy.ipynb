### Creating numpy arrays

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([[1.1,2.4],[3.6,4.5]])
arr3 = np.array([[[1,2,3,4,5],[5,6,7,8,9],[9,1,2,3,4]]])
arr4 = np.array([[[3,4,5],[2,4,5]],
                 [[4,5,6],[7,8,9]],
                 [[5,3,2] ,[3,6,3]]])
print(arr1)
print(arr2)
print(arr3)
print(arr4)
```

### Mathematical functions

```python
a1 = [7,8,9]
a2 = [4,5,6]

print(f"addition: {np.add(a1 ,a2)}")
print(f"subtraction: {np.subtract(a1 ,a2)}")
print(f"multiplication: {np.multiply(a1 ,a2)}")
print(f"division: {np.divide(a1 ,a2)}")
print(f"power: {np.pow(a1 ,2)}")
print(f"sqrt: {np.sqrt(a2)}")

print(f"sum: {np.sum(a1)}")#addition of all elements in one array



```

### statistical function

```python
a1 = [2,4,65,23,90,-5,5]

print(f'mean:{np.mean(a1)}')
print(f'median:{np.median(a1)}')
print(f'standard deviation:{np.std(a1)}')
print(f'variance:{np.var(a1)}')
print(f'minium :{np.min(a1)}')
print(f'maxium:{np.max(a1)}')
print(f'percentile:{np.percentile(a1,33)}')

```

### comparison & logical function

```python
arr = np.array([5,10,15,20,25])

print(np.where(arr%2 == 0 , arr , 0))
print(np.any(arr%2 == 0)) # return true if any one element satisfy condition
print(np.all(arr%2 == 0)) # return true if all element satisfy condition
print(np.logical_and(arr> 8, arr < 18))
print(np.logical_or(arr > 8, arr < 18))
```

### View of Array in NumPy

```python

arr = np.array([2, 4, 6, 8, 10])
v = arr.view()

print("Original ID:", id(arr))
print("View ID:    ", id(v))

arr[0] = 12 

print("Original array:", arr)
print("View array:    ", v)
```

### Copy of Array in NumPy

```python

arr = np.array([2, 4, 6, 8, 10])
c = arr.copy()

print("Original ID:", id(arr))
print("Copy ID:    ", id(c))

arr[0] = 12
print("Original array:", arr)
print("Copy array:    ", c)
```

flatten() â€“ Convert to 1D (copy) . 
Converts array to 1D array and returns a copy.

```python
a = np.array([[1, 2],
              [3, 4]])

b = a.flatten()
print(b)
```

ravel() - same as flatten but it return view not copy

```python
a = np.array([[1, 2],
              [3, 4]])

b = a.ravel()
print(b)
```

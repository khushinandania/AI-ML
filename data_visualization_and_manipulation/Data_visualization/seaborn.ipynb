# lineplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x1=[1,2,3,4,5]
y1=[4,7,3,1,9]
x=pd.DataFrame({"x1":x1,"y1":y1})
sns.lineplot(x="x1",y="y1",data=x)
plt.show()
```

```python
y=sns.load_dataset("penguins")
y
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y,hue="sex",size=10,style="sex",palette='Accent',markers=["o",">"],dashes=False)
plt.grid()
plt.show()
```

# Histogram

```python
y=sns.load_dataset("penguins")
sns.displot(y["bill_depth_mm"],bins=[14,15,16,17,18,19,20],kde=True,rug=True,color="g")
plt.show()
```

# Bar plot

```python
y=sns.load_dataset("penguins")
sns.set(style="darkgrid")
sns.barplot(x="island" , y="bill_length_mm" , data=y ,hue="sex",saturation=0.2 ,palette="Accent",capsize=0.2,
            err_kws={'color': 'y','linewidth': '2'},hue_order=["Female","Male"],alpha=0.7)
plt.show()
```

# Scatter Plot

```python
y=sns.load_dataset("penguins")
m={"Male":"o" , "Female":"s"}
sns.scatterplot(x="flipper_length_mm" , y="bill_length_mm" , data=y ,hue="sex",style="sex",size="sex",sizes=(60,40),palette="Accent",
           alpha=0.7,markers=m)
plt.show()
```

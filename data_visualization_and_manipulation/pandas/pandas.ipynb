#### Handling missing value

```python
df.dropna() # by default axis =0 drop rows which has nan values
```

```python
df.dropna(axis=1)  # drop that column which has missing value
```

```python
df.dropna(how ='any') # DROP that rows which has any one missing value
```

```python
df.loc[len(df)] = np.nan
df
```

```python
df.dropna(how='all') #drop that rows which has all missing values
```

#### Fill missing values

```python
df.fillna('0')
```

```python
df.bfill(axis=0)
```

```python
df.ffill(axis=1)
```
